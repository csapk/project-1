```python
outlier_criteria = (df < (q1-1.5*iqr)) | (df > (q3+1.5*iqr))
outlier_criteria
```

```python
outlier_criteria.sum()
```

```python
import pandas as pd
import numpy as np 

import seaborn as sns 
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler
```

```python
# define category and numeric

nums = mall[['Islam', 'Kristen', 'Katolik', 'Hindu', 'Budha', 'Konghucu', 'Kepercayaan']]
cat = mall[['Kecamatan']]
```

```python
cat
```

```python
# Optimasi K-Means dengan metode elbow untuk menentukan jumlah klaster yang tepat
from sklearn.cluster import KMeans 
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42) 
    kmeans.fit(nums)
    wcss.append(kmeans.inertia_) 

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Cluster Number')
plt.ylabel('WCSS')
plt.show()
```

Berdasarkan metode elbow, grafik berbentuk elbow terdapat pada cluster 2 dan 3, maka dapat disimpulkan pembagian klaster 2 atau 3 adalah yang paling optimum.

```python
# Proses K-Means Clustering
km_elbow = KMeans(n_clusters = 3, 
                     init = 'k-means++', 
                     max_iter = 100, 
                     n_init = 10, 
                     random_state = 42)
y_kmeans = km_elbow.fit_predict(nums.values)
```

```python
# concate dengan dataframe hasil copy tadi, yaitu mall

mall['Cluster'] = y_kmeans
```

```python
mall.head()
```

```python
average = mall.groupby(['Cluster'], as_index = False).mean()
average
```

Tabel di atas menunjukkan cluster beserta rata-rata jumlah penganut agama.

```python
features = ['Islam', 'Kristen', 'Katolik', 'Hindu', 'Budha', 'Konghucu', 'Kepercayaan']
```

## Hasil Perbandingan

Dilakukan analisis terhadap scatterplot hasil dari metode K-Means Clustering dan scatterplot hasil dari metode Hierarchical Clustering. Berdasarkan analisis tersebut, diperoleh bahwa penggunaan metode K-Means Clustering lebih sesuai dibandingkan penggunaan metode Hierarchical Clustering. Hal ini dikarenakan hasil clustering dari metode K-Means menunjukkan pembagian objek data ke dalam kelompok tidak saling overlap, sehingga setiap data berada tepat di satu cluster. Sedangkan hasil clustering dari metode Hierarchical kelompok datanya saling overlap dan tidak tepat di satu cluster.
